```python
from __future__ import print_function, division
import os
import re
import sys
import glob

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

print('sys_version', sys.version.replace('\n', ''))
print('virtual_env', os.environ.get('VIRTUAL_ENV', 'None'))
print('pwd', os.getcwd())
print('np ', np.__version__)
```

```python
DEC_FILE = "./izlaz-dec.txt"
OBJ_FILE = "./izlaz-obj.txt"

def read_data(path):
    with open(path, 'r') as fin:
        def line_to_np(line):
            line = line.strip()[1:-1] # remove []
            str_elems = re.split('\s*,\s*', line) #split
            return np.array([float(x) for x in str_elems])
        lines = [line_to_np(line) for line in fin]
    return np.array(lines)

Xs = read_data(DEC_FILE)
ys = read_data(OBJ_FILE)

print(len(Xs), len(ys))
```

```python
x1, x2 = zip(*Xs)
plt.plot(x1, x2, 'o')
```

```python
y1, y2 = zip(*ys)
plt.plot(y1, y2, 'o')
plt.ylim(0, 10)
plt.xlim(0.1, 1)
```
